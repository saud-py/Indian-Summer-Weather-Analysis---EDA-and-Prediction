```python
df = summer_data.drop(['Date', 'sunrise', 'sunset', 'description'], axis = 1)
```

```python
#The quantile() is numpy function which takes an array and number say 0 and 1 and returns the qth quantile value 
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
df = df[~((df<(Q1-1.5*IQR)))]
#The IQR is the inter-quantile-range is the difference between the 75th and 25th percentile of the data
```

```python
df.humidity = np.sqrt(df.humidity)
df.windspeed = np.sqrt(df.windspeed)
```

```python
sns.set(style = "darkgrid")
fig, axs = plt.subplots(2, 2, figsize = (10, 8))
sns.histplot(data = df, x = "humidity", kde = True, ax = axs[0, 0], color = 'green')
sns.histplot(data = df, x = "tempmax", kde = True, ax = axs[0, 1], color = 'red')
sns.histplot(data = df, x = "tempmin", kde = True, ax = axs[1, 0], color = 'skyblue')
sns.histplot(data = df, x = "windspeed", kde = True, ax = axs[1, 1], color = 'orange')
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lc = LabelEncoder()
df1["conditions"] = lc.fit_transform(df1["conditions"])
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['tempmax'] = df1['tempmax'].fillna(0)
df1['tempmin'] = df1['tempmin'].fillna(0)
df1['temp'] = df1['temp'].fillna(0)
df1['feelslikemax'] = df1['feelslikemax'].fillna(0)
df1['feelslikemin'] = df1['feelslikemin'].fillna(0)
df1['dew'] = df1['dew'].fillna(0)
df1['windspeed'] = df1['windspeed'].fillna(0)
df1['winddir'] = df1['winddir'].fillna(0)
```

```python
#The Below snippet won't run unless you the clean the df1, because it contains null which cannot be converted to int from astype method 
#The df1 is cleaned by filling the Nan values to 0 by using the fillna(0)
x = ((df1.loc[:,df1.columns != "conditions"]).astype(int)).values[:,0:]
y = df1["conditions"].values
```

```python
x_train, x_test,y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
print("KNN Accuracy:{:.2f}%".format(knn.score(x_test, y_test)*100))
```

```python
svm = SVC()
svm.fit(x_train, y_train)
print("SVM Accuracy : {:.2f}%".format(svm.score(x_test, y_test)*100))
```

```python
gbc=GradientBoostingClassifier(subsample=0.5,n_estimators=450,max_depth=5,max_leaf_nodes = 5)
gbc.fit(x_train,y_train)
print("Gradient Boosting Accuracy:{:.2f}%".format(gbc.score(x_test,y_test)*100))
```
